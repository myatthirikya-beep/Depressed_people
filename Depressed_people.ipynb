```python
import matplotlib.pyplot as plt
df['Gender'].value_counts().plot.pie(
    autopct =  '%1.1f%%',
    startangle = 0,
    shadow = True,
    figsize = (8,8)
)
plt.title('Gender Distribution')
plt.ylabel(" ")
plt.show()
```

```python
df['Age'].value_counts().plot.hist(
    figsize = (5,4),
    bins = 7,
    color = 'blue',
    edgecolor = 'black'
)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
df['Academic Pressure'].value_counts().sort_index().plot.bar(
    figsize = (7,4),
    color = 'green',
    edgecolor = 'black',

)
plt.title('Academic Pressure Distribution')
plt.xlabel('Academic Pressure')
plt.ylabel('Count')
plt.legend(['Academic Pressure'])
plt.show()
```

```python
df['Study Satisfaction'].value_counts().sort_index().plot.bar(
    figsize = (5,4),
    color = 'skyblue',
)
plt.title('Study Satifaction Distribution')
plt.xlabel('Study Satisfaction Rate')
plt.ylabel('Count')
plt.legend(['Study Satisfaction'])
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
sns.histplot(
    df['Sleep Duration'],
    bins=7,
    color='gold',
    edgecolor='black'
)
plt.title('Sleep Duration Distribution', fontsize=13, weight='bold')
plt.xlabel('Sleep Duration (hours)', fontsize=11)
plt.ylabel('Count', fontsize=11)
sns.despine()
plt.tight_layout()
plt.show()
```

```python
pd.crosstab(df['Dietary Habits'], df['Depression']).plot(
    kind='bar',
    stacked=True,
    color=['lightblue','salmon'],
    figsize=(6,4),
    edgecolor='black'
)
plt.title('Comparison: Dietary Habits vs Depression')
plt.xlabel('Dietary Habits')
plt.ylabel('Number of Individuals')
plt.legend(title='Depression')
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.boxplot(
    x='Depression',
    y='Sleep Duration',
    data=df,
    palette=['#F48FB1','#81D4FA'],
    width=0.5,
    boxprops = dict(alpha=0.8)
)


plt.title('Sleep Duration Distribution by Depression Status', fontsize=13, weight='bold')
plt.xlabel('Depression Status', fontsize=11)
plt.ylabel('Sleep Duration (hours)', fontsize=11)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
df['Dietary Habits'].value_counts().plot.pie(
    figsize = (5,5),
    autopct = '%1.1f%%',
    startangle = 90,
    shadow = True,
    colors  = ['#ff9999','#66b3ff', 'yellow' ], # blue and pink
    explode = (0.1,0.1,0.1),
  )

plt.title('Dietary Habits Distribution')
plt.ylabel(' ')
plt.show()
```

```python

pd.crosstab(df['Have you ever had suicidal thoughts ?'],df['Family History of Mental Illness']).plot(
    kind = 'bar',
    figsize = (10,6),
    color = ['red','blue'],
    edgecolor = 'black',
    stacked=True)
plt.title('Percentage Comparison: Family History vs Suicidal Thoughts')
plt.ylabel('Percentage (%)')
plt.show()
```

```python
avg = df.groupby('Study Satisfaction')['Study Hours'].mean().reset_index()
sns.barplot(x='Study Satisfaction', y='Study Hours', data=avg, palette='coolwarm')
plt.title('Average Study Hours by Satisfaction Level')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Get value counts for both columns, sorted by index (1–5)
financial_counts = df['Financial Stress'].value_counts().sort_index()
academic_counts = df['Academic Pressure'].value_counts().sort_index()

# Create index positions
index = np.arange(len(financial_counts))  # should be 5 if scales are 1–5
bar_width = 0.35
opacity = 0.8

# Create figure
plt.figure(figsize=(10,4))

# Plot bars side-by-side
plt.bar(index, financial_counts, bar_width, alpha=opacity, color='steelblue', label='Financial Stress')
plt.bar(index + bar_width, academic_counts, bar_width, alpha=opacity, color='mediumseagreen', label='Academic Pressure')

# Add labels and title
plt.xlabel('Rating (1–5)')
plt.ylabel('Count')
plt.title('Comparison between Financial Stress and Academic Pressure')
plt.xticks(index + bar_width/2, financial_counts.index)  # Center x-ticks
plt.legend()

# Style tweaks
plt.tight_layout()
plt.show()
```
